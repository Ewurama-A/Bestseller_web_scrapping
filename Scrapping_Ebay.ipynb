# Webscrapping E-Bay
This is an exercise to obtain data from the popular e-commerce site E-bay. The category selected is PC laptops from different brands ranging from dell to lenovo as well as emerging brands. The product name, brand and  price range. Further cleaning and manipulation using pandas and siuba will be done in another notebook. The data is 
from April 2023.

```python
# Using requests-html's Async session 

from requests_html import AsyncHTMLSession

# Creating headers to emulate client-servers

headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:66.0) Gecko/20100101 Firefox/66.0", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

asession = AsyncHTMLSession()

```

```python
# The url from ebay and setting page variable

page_no = 1

url = f"https://www.ebay.com/b/PC-Laptops-Netbooks/177/bn_317584?_pgn={page_no}"
```

```python
# Request to get url using headers and Async

get_session = await asession.get(url, headers = headers)
```

```python
get_session.status_code
```

```python
#Get the product name

product = []

while page_no <= 10:
    product_path = get_session.html.find('h3.s-item__title')
    [product.append(p.text) for p in product_path]
    page_no+=1
    
print(product)
print(len(product))
```

```python
#Get the prices

price = []

while page_no<= 10:
    price_path = get_session.html.find('span.s-item__price')
    [price.append(x.text.replace("$","")) for x in price_path]
    page_no+=1
    
print(price)

print(len(price))
```

```python
# Get brand
# Brand name is the first word in product name

Brand = []

for b in product:
    if "Dell" in b:
        Brand.append("Dell")
    elif "Lenovo" in b:
        Brand.append("Lenovo")
    elif "Samsung" in b:
        Brand.append("Samsung")
    elif "HP" in b:
        Brand.append("HP")
    elif "Acer" in b:
        Brand.append("Acer")
    elif "Microsoft" in b:
        Brand.append("Microsoft")
    elif "Panasonic" in b:
        Brand.append("Panasonic")
    else: Brand.append('Other')

print(Brand)
```
