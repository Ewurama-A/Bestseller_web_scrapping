### My first webcrapping program

This program is attempting to parse through audible.com and obtain the bestsellers on their website.  Packages such as requests, httpx and parsel will be used to get the titles, ratings, languages and release date of the given books. 
Commented code will show what is happening

```python
import this
```

```python
import requests as rq


url= "https://www.audible.com/adblbestsellers?ref=a_hp_t1_navTop_pl0cg1c0r0&pf_rd_p=c592ea51-fd36-4dc9-b9aff665ee88670b&pf_rd_r=0Z335GYQABPDNWE8R84N"


#mimicking a web broswer request to the server
headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36(KHTML, like Gecko) Chrome/87.0.4280.66 Safari/537.36",
"Accept-Language": "en-US,en;q=0.8",
"Accept-Encoding": "gzip, deflate"}

#the request to the server
response = rq.get(url, headers=headers)

#checking if the request was successful
if response.status_code == 200:
 print("Successful request")
else:
 print("Request failed")

#review content
response.content

```

```python
#parsel selector to choose specific content

from parsel import Selector

html = response.content

s = Selector(str(html))
```

```python
#Using parsel and requests to get titles

titles = s.xpath("//h3/a/text()").getall()
# Or use css selectors: s.css('h3>a::text').extract()

"""Before list comprehension, loop through content once to get background of raw data, then cleaning can continue 
E.g for t in titles: print(t)"""

print(titles)


```

```python

release_date=s.xpath('//li[contains(@class,"releaseDateLabel")]/span/text()').getall()
release_date2 =[x.split()[2].replace("\n", "") for x in release_date]
print(release_date2)
```

```python

authors = s.xpath("//li[contains(@class, 'authorLabel')]/span/a/text()").getall()
print(type(authors))
authors[-4]=['Jinger Duggar Vuolo','Corey Williams']

print(authors)
#authors2 = authors.del
#authors2 = [print(a) for a in authors[:-3]]
```

```python

language_type=s.xpath('//li[contains(@class,"languageLabel")]/span/text()').getall()

language_type2 = [l.split()[1] for l in language_type]

print(language_type2)
#using spitlines
#def func(value):
    #return ''.join(value.splitlines())

#print(func(str(language_type)))

#another way to clean the data using regex
#language_type=re.sub('\\n','',str(language_type))


    
```

```python
#narration runtime

length = s.xpath('//li[contains(@class,"runtimeLabel")]/span/text()').getall()

length2 = [l.split(":")[1].replace("\n", "") for l in length]

print(length2)

#https://www.bookdepository.com/bestsellers
```

```python
#Using httpx which is another way to get content from a url instead of requests

import httpx

# function to call if httpx is preferred
def get_html(url): 
    html2 = httpx.get(url).content
    return html2
```

```python
# Call  function

html2 = get_html(url)

s2 = Selector(str(html2))
```

```python
# ratings

ratings = s2.xpath('//span[has-class( "bc-pub-offscreen")]/text()').extract()
```

```python
ratings3 = [float(x.split(" ")[0]) for x in ratings[:20]]

ratings3[2] = 4.5
ratings3[5] = 4.5
ratings3[7] = 4.5
ratings3[13] = 4.5
ratings3[14] = 4.5
ratings3[15] = 4.5
ratings3[16] = 4.5
ratings3[5] = 4.0

print(ratings3)
```

```python
#check length to ensure accuracy

print(len(ratings3))
print(len(length2))
print(len(release_date2))
print(len(authors))
print(len(titles))
print(len(language_type2))
```

```python
import pandas as pd

#creating a dataframe using dictionaries
pdframe= {'title':titles,'author':authors, 'rating':ratings3, 'language': language_type2, 'release_date':release_date2,'runtime':length2}
#print(pdframe)

audible_data = pd.DataFrame(pdframe)


#check whether dataframe was created

audible_data.head()

#audible_data['author'], checking author column for accuracy
#audible_data['title']=pd.Series(titles), i.e another way to create columns
```

```python
#saving dataframe to a csv file
audible_data.to_csv("audible_bestsellers.csv")
```
